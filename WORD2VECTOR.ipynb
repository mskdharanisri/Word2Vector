```python
corpus_raw = 'Hello World . Hello hi . hello queen'
corpus_raw = corpus_raw.lower()
```

```python
words = []
for word in corpus_raw.split():
    if word != '.':
        words.append(word)
words = set(words)
word2int = {}
int2word = {}
vocab_size = len(words)
for i,word in enumerate(words):
    word2int[word] = i
    int2word[i] = word
```

```python
print(words)
print(word2int)
```

```python
raw_sentences = corpus_raw.split('.')
sentences = []
for sentence in raw_sentences:
    sentences.append(sentence.split())
```

```python
data = []
WINDOW_SIZE = 2
for sentence in sentences:
    for word_index, word in enumerate(sentence):
        for nb_word in sentence[max(word_index - WINDOW_SIZE, 0) : min(word_index + WINDOW_SIZE, len(sentence)) + 1] :
            if nb_word != word:
                data.append([word, nb_word])
```

```python
def to_one_hot(data_point_index, vocab_size):
    temp = np.zeros(vocab_size)
    temp[data_point_index] = 1
    return temp
x_train = []
y_train = []
for data_word in data:
    x_train.append(to_one_hot(word2int[ data_word[0] ], vocab_size))
    y_train.append(to_one_hot(word2int[ data_word[1] ], vocab_size))
x_train = np.asarray(x_train)
y_train = np.asarray(y_train)
print(y_train)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Define your model using PyTorch
class WordEmbeddingModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim):
        super(WordEmbeddingModel, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.fc = nn.Linear(embedding_dim, vocab_size)

    def forward(self, x):
        embedded = self.embedding(x)
        output = self.fc(embedded)
        return torch.nn.functional.softmax(output, dim=1)

# Define your model and loss function
vocab_size =5  # Define your vocabulary size
embedding_dim = 100  # Define the embedding dimension
model = WordEmbeddingModel(vocab_size, embedding_dim)
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)

# Replace these placeholders with your actual training data
#x_train = torch.Tensor(your_x_train_data).long()
#y_train = torch.Tensor(your_y_train_data).long()
x_train = np.random.randint(2, size=(2, 5))  # Example input data
y_train = np.random.randint(2, size=(2, 5))
x_train = torch.Tensor([[0, 0, 1, 0, 1],
 [1 ,1 ,0 ,0 ,0]]).long()
y_train = torch.Tensor([[1 ,1, 1, 1, 0],
 [0 ,1 ,1 ,1 ,1]]).long()
# Training loop
n_iters = 100
for i in range(n_iters):
    optimizer.zero_grad()
    outputs = model(x_train)
    loss = criterion(outputs, y_train)
    loss.backward()
    optimizer.step()
    print(loss.item())
```
